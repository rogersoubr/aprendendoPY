```python
import pandas as pd #coloca pd como nome da lib pandas
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df
```

```python
df.head(10)#mostra o cabecalho. neste caso, 10 primeiras linhas
```

```python
df.tail(10)#mostra o final. neste caso, 10 ultimas linhas
```

```python
df.info()#mostra as informações da coluna
```

```python
#renomeando as colunas. passa um dicionário com o nome que esta e o nome que quer
df = df.rename(columns={
    'PassengerId':'ID',
    'Survived': 'SOBREV',
    'Pclass': 'Classe',
    'Name':'NOME',
    'Age':'IDADE',
    'Sex':'SEXO',
    'SibSp':'N_ADULTOS',
    'Parch':'N_CRIANCAS',
    'Ticket':'BILHETE',
    'Fare':'TAXA',
    'Cabin':'CABINE',
    'Embarked':'LOCAL',
})
df.head()
```

```python
df.SOBREV = df.SOBREV.astype(bool)#muda o tipo da coluna SOBREV
df['FALECEU'] = df.SOBREV#cria outra coluna 
df.head()
```

```python
df.TAXA = df.TAXA * 5.15#muda para real
df['TOTAL_PESSOAS'] = df.N_ADULTOS + df.N_CRIANCAS + 1
df['TAXA_PESSOA'] = df.TAXA / df.TOTAL_PESSOAS
df.head()
```

```python
#torca valores
df.NOME = df.NOME.str.lower();
df.nome = df.NOME.replace('mrs','sra')
df.nome = df.NOME.replace('mr','sr')
df.nome = df.NOME.replace('miss','srta')
df.head()
```

```python
#muda mapeando male para masc
df.SEXO = df.SEXO.map({'male':'masc', 'female':'fem'})
df.head()
```

```python
df.CABINE = df.CABINE.fillna('PORÃO')#tratando valores nulos
df.head()
```

```python
def define_tipo(total_pessoas):
    if total_pessoas == 1:
        return 'individual'
    if total_pessoas == 2:
        return'casal'
    if total_pessoas >= 3:
        return 'familia'

df['TIPO'] = df.TOTAL_PESSOAS.apply(define_tipo)#adiciona uma nova tabela que tem uma coluna com uma funcao aplicada
df.head() 
#desk, spark são melhores que pandas nisso
```

```python
df.sort_values(['TAXA','TOTAL_PESSOAS'])#mostra em crescente as taxas por total pessoas
#vai ordenas primeiro por taxa. Depois por pessoas
```

# Agregando dados

```python
df.groupby('SEXO').mean(numeric_only=True)#agrupa por femininos e masculinos e faz a média das tabelas numerticas
```

```python
df.groupby(['SEXO', 'Classe']).SOBREV.mean(numeric_only=True) #vai mostrar que a a classe femina baixo morrei mais
```

```python
df.groupby('TIPO').agg({'IDADE': 'mean', 'TAXA':'sum', 'SOBREV':'mean', 'BILHETE':'sum'})
```
